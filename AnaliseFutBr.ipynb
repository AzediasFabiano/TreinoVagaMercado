###Analise Brasileirão Rodada 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

- Dados coletados do SofaScore 

```python
rodada1 = pd.DataFrame({
    "Times": ["AtleticoGO", "Flamengo", "Palmeiras", "Ceara", "Fluminense", "Santos", "Coritiba", "Goias", "AtleticoMG",
              "Internacional", "Botafogo", "Corinthians", "Fortaleza", "Cuiaba", "Avai", "AmericaMG", "Sao Paulo",
              "AthleticoPR", "Juventude", "RBBragantino"],
    "Gols":           [1, 1, 2, 3, 0, 0, 3, 0, 2, 0, 1, 3, 0, 1, 1, 0, 4, 0, 2, 2],
    "Assistencias":   [1, 1, 1, 0, 0, 0, 2, 0, 2, 0, 0, 3, 0, 0, 0, 0, 3, 0, 0, 1],
    "FinalizaçaoGol": [3, 3, 4, 9, 6, 1, 9, 0, 5, 1, 4, 8, 5, 4, 2, 1, 8, 0, 7, 2],
    "GrandesChances": [6, 4, 3, 2, 3, 0, 2, 1, 7, 1, 2, 5, 2, 0, 0, 1, 1, 0, 2, 1],
})
```

```python
plt.figure(figsize=(18, 6))
plt.plot(rodada1.loc[:, "Times"], rodada1.loc[:, "Assistencias"], color='orange', label='Assistencias')
plt.plot(rodada1.loc[:, "Times"], rodada1.loc[:, "FinalizaçaoGol"], color='blue', label='FinalizaçaoGol')
plt.plot(rodada1.loc[:, "Times"], rodada1.loc[:, "GrandesChances"], color='black', label='GrandesChances')
plt.plot(rodada1.loc[:, "Times"], rodada1.loc[:, "Gols"], color='red', label='Gols')
plt.xticks(rotation=40, fontsize=16, fontweight='bold')
plt.yticks(np.arange(0, 10, 1), fontsize=12, fontweight='bold')
plt.title('Rodada 1 do Brasileirão: Estatística de: Gols x Assistências x Finalizações no gol x Grandes chances', fontsize=16, fontweight='bold', fontstyle='italic')
plt.legend()
```

-Correlação entre os indices

```python
corr_mat = rodada1[['Gols', 'Assistencias', 'FinalizaçaoGol', 'GrandesChances']].corr()
mask = np.array(corr_mat)
mask[np.tril_indices_from(mask)] = False
sns.heatmap(corr_mat, mask=mask, square=True, annot=True)
plt.title('Correlação dos índices', fontsize=16, fontweight='bold', fontstyle='italic')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.show()
```

###Área de Automatização

```python
listaTimes = []
listaGols = []
listaAssistencia = []
listaFinalizaçaoGol = []
listaGrandesChances = []
```

```python
while True:
      times = input("Nome do time: ")
      gols = int(input("Gols marcados: "))
      assist = int(input("Assistências: "))
      finalização = int(input("Chutes no gol: "))
      chances = int(input("Grandes chances + bola na trave: "))
      botão = input("Digite ok p/ parar ou Clica em Concluir: ").lower()
      listaTimes.append(times)
      listaGols.append(gols)
      listaAssistencia.append(assist)
      listaGrandesChances.append(chances)
      listaFinalizaçaoGol.append(finalização)
      if botão == 'ok':
        print('Obrigado por ajudar.')
        break
      elif len(listaTimes) == 20:
        print('Concluído.')
        break
      else:
        continue
rodada2 = pd.DataFrame ({
      "Times" : listaTimes,
      "Gols" : listaGols,
      "Assistências" : listaAssistencia,
      "Finalização Gol" : listaFinalizaçaoGol,
      "Grandes chances" : listaGrandesChances      
})
```

```python
#Criando um arquivo csv com o dataframe
rodada2.to_csv('rodada2_BrTeste2.csv')
```
